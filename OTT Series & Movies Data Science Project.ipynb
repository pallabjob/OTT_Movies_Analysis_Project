```python
# Univariate Analysis
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Rating Distribution
sns.histplot(df['rating'], bins=20, kde=True)
plt.title("Distribution of Ratings")
plt.show()
```

This distribution peaks around 7 to 8. There are some movies in the 5 to 7 range, the highest concentration is slightly above average (around 7-7.5). So most of the movies are above average rather than just average.

There is a strong presence of ratings between 7 and 9 The right side (7.5–9) has many bars with high counts, meaning a significant number of movies are well-rated.
However, extremely high ratings (above 9) are rare.
                                                                                                                                    


```python
# Genre Analysis
df['genre'].value_counts().head(10).plot(kind='bar')
plt.title("Top Genres")
plt.show()
```

Yes, OTT platforms are heavily dominated by Drama, Sci-Fi, and Crime content.
This suggests:

Viewers prefer story-driven and emotionally engaging content (Drama)
High demand for imaginative/tech-based content (Sci-Fi)
Strong interest in suspense and investigation (Crime)

```python
# language Distribution
df['language'].value_counts().head(10).plot(kind='bar')
plt.title('Top Languages')
plt.show()
```

According to Language distribution German Language is highest Distribution after that English is very nearly to German Language. Third domination language is Spanish and fourth domination of language is Hindi. Spanish and Hindi are lower distribution than German and English but higher distribution than Korean language. 

# Bivariate Analysis (relationships)

```python
# Rating vs Genre
df.groupby('genre')['rating'].mean().sort_values(ascending=False).head(10)
```

According to Rating vs Genre relationships Action is the highest ratings. Second highest ratings is Sci-Fi and Lowest ratings is Thriller.

```python
# Votes vs Rating
sns.scatterplot(x='votes', y='rating', data=df)
plt.title('Votes vs Ratings')
plt.show()
```

“Based on scatter plot analysis, there is no strong correlation between votes and ratings. Ratings are distributed across all vote levels, indicating that popularity does not guarantee higher user satisfaction.”

```python
# year-wise Trend
df['release_year'].value_counts().sort_index().plot()
plt.title("Movies Released Over Time")
plt.show()
```

Growth of OTT content is enhancing from 2016 to 2024. In 2016 the growth is starting and in 2024 the growth of OTT content is enhancing tremendously.

# Feature Engineering

```python
df['popularity']=df['rating']*df['votes']
df['description_length']=df['description'].apply(lambda x:len(str(x)))
```

```python
df.head()
```

# NLP EDA

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words='english')
word_matrix = cv.fit_transform(df['description'].astype(str))

word_counts = word_matrix.toarray().sum(axis=0)
words = cv.get_feature_names_out()

word_freq = pd.DataFrame({'word': words, 'count': word_counts})
word_freq.sort_values(by='count', ascending=False).head(20)
```

Engaging, Featuring, Intense, Storytelling, Characters Content dominates

# Correlation Heatmap

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

Weighted_rating is the strongest predictor of movie ratings.
Audience-related metrics such as popularity, engagement, and trending score show a mild positive influence.
Features like release year, duration, and description length do not significantly affect ratings.
